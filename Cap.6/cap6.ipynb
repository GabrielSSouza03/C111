```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

PLOT

```python
# Descreve as alterações nos dados ao longo do tempo 
x = np.array([1,2,3,4])
y = x*2 # eixo y do gráfico 1
y2 = x**2 # eixo y do gráfico 2

#utilizando o conceito de subplot
plt.subplot(1,2,1)
plt.title('Função Linear')
plt.plot(x,y,'s:r')

plt.subplot(1,2,2)
plt.title('Função Exponencial')
plt.plot(x,y2,'o-b')




#plt.xlabel('valores de X')
#plt.ylabel('valores de Y')

#plt.plot(x,y, 's:r',x,y2,'o-b', linewidth=3, markersize=20)
```

GRÁFICO EM BARRAS

```python
#importanto o dataset
df= pd.read_csv('paises.csv',delimiter=';')

#extraindo as colunas 
print(df.columns)

#Extraindo os 5 maiores GDP do dataset 
dfBiggestGDP = df.nlargest(5, 'GDP ($ per capita)')

#extraindo o nome destes paises 
eixo_x = dfBiggestGDP['Country']
eixo_y = dfBiggestGDP['GDP ($ per capita)']

#traçando o graáfico em barrasa 
plt.bar(eixo_x, eixo_y, color='green')

#print(dfBiggestGDP)
```

GRÁFICO EM TORTA

```python
print(df.columns)

#Extraindo países que não tem costa marítma 
dfNoCoast = df[df['Coastline (coast/area ratio)'] == 0]
withCoast = len(df)- len(dfNoCoast)
withoutCoast = len(dfNoCoast)

#traçando o gráfico em torta 
plt.pie(x=[withCoast, withoutCoast], labels=['% Paises com Costa', '% Paises sem Costa'],autopct ='%1.1f%%' )
```
